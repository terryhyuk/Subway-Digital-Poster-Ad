```python
subway = pd.read_csv('../data/역별 일별 시간대별 승하차인원 정보 23-06_24_05.csv')
subway.head()
```

```python
subway['date']=pd.to_datetime(subway['날짜'])
subway.head()
```

```python
subway['date']=pd.to_datetime(subway['date'])
days=["월", "화", "수", "목", "금", "토", "일"]

subway['day']=subway['date'].dt.weekday.map(dict(enumerate(days)))
```

```python
subway.head()
```

```python
subway.info()
```

```python
colnames = subway.columns[5:25]
```

```python
subway_sort = subway.pivot_table(
    colnames,
    index=["date","day","역명","구분"],
    aggfunc=sum
)
```
